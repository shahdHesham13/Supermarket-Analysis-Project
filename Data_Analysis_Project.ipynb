## Analysis & Visualization

### Asking bussiness questions to understand the data better and visualize for clearer more inclusive look ino the data values

## Which distributer has the highest target?

```python
pvt_table = pd.pivot_table(
    finalTarget,
    values='Target',
    index='Distributor',
    aggfunc='max'
)
pvt_table
```

```python
bw = 0.2
x1 = pvt_table.index
y1 = pvt_table['Target']

max_index = y1.idxmax()
colors = ['blue' if i != max_index else 'darkblue' for i in y1.index]

x2 = [i + bw for i in range(len(pvt_table.index))]
y2 = pvt_table['Target']

plt.bar(x1, y1, width = bw, color=colors, label = 'Target')

plt.legend()
```

## What target each zone have?

```python
zonepvt_table = pd.pivot_table(
    finalTarget,
    values='Target',
    index='ZONE',
    aggfunc='sum'
)
zonepvt_table
```

```python
explode = [0] * len(zonepvt_table)
explode[0] = 0.25

plt.pie(zonepvt_table['Target'],
        labels=zonepvt_table.index,
        autopct='%1.0f%%',
        explode=explode,
        colors=['lightgreen', 'pink', 'skyblue'][:len(zonepvt_table)]);
```

## Visualizing Target distribution across Zones

```python
pvt_table = pd.pivot_table(
    finalTarget,
    values='Target',
    index='Distributor',
    aggfunc='max'
)
pvt_table
```

```python
zones = pvt_table.index.tolist()
targets = pvt_table['Target'].tolist()

fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)
axs[0].bar(zones, targets, color='skyblue')
axs[1].scatter(zones, targets, color='green')
axs[2].plot(zones, targets, marker='o', color='orange')

axs[0].set_title('Bar Plot')
axs[1].set_title('Scatter Plot')
axs[2].set_title('Line Plot')
fig.suptitle('Sum of Targets by Zone')

plt.show()
```

## How do target values vary across different zones and distributors?

```python
pivot = pd.pivot_table(
    finalTarget,
    values='Target',
    index=['ZONE', 'Distributor'],
    columns='Role',
    aggfunc=['sum', 'mean']
)
pivot
```

# Warehouses & Outlets Data

```python
Pathware = '/content/drive/MyDrive/carerhaPractice/Data/Data Analysis Project - Practice/Warehouses.xlsx'
Pathout = '/content/drive/MyDrive/carerhaPractice/Data/Data Analysis Project - Practice/Outlets.csv'

warehouseData = pd.read_excel(Pathware)
outletData = pd.read_csv(Pathout)
```

```python
warehouseData.head(3)
```

```python
outletData.head(3)
```

```python
outletsFinal = pd.merge(warehouseData, outletData, on='Warehouse Name', how='left')
outletsFinal
```

```python
outletsFinal.info()
```

```python
outletsFinal['Code'] = outletsFinal['Code'].astype('object')
outletsFinal['OutletId'] = outletsFinal['OutletId'].astype('object')
outletsFinal['Employee_Code'] = outletsFinal['Employee_Code'].astype('object')
outletsFinal['Warehouse Code'] = outletsFinal['Warehouse Code'].astype('object')
```

```python
outletsFinal['Outlet Class'].unique()
```

```python
outletsFinal['Outlet Type'].unique()
```

## There is two different value with the same meaning, will change to standardize the spelling.

```python
outletsFinal['Outlet Type'] = outletsFinal['Outlet Type'].replace({'تجزئه': 'تجزئة'})
```

```python
outletsFinal['Outlet Type'].unique()
```

```python
outletsFinal.head(3)
```

## Analysis & Visualization

### Asking bussiness questions to understand the data better and visualize for clearer more inclusive look ino the data values

## How many outlets of each class each region have?

```python
# Pivot table
pvt_outlets = pd.pivot_table(
    outletsFinal,
    values='OutletId',
    index='Region',
    columns='Outlet Class',
    aggfunc='count',
    fill_value=0
)

# Visualization
pvt_outlets.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Total Outlets by Region and Outlet Class')
plt.xlabel('Region')
plt.ylabel('Number of Outlets')
plt.legend(title='Outlet Class')

plt.show()
```

## How many outlets of each type each region have?

```python
# Pivot table
pvt_outlets = pd.pivot_table(
    outletsFinal,
    values='OutletId',
    index='Region',
    columns='Outlet Type',
    aggfunc='count',
    fill_value=0
)

# Visualization
pvt_outlets.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='summer')
plt.title('Total Outlets by Region and Outlet Class')
plt.xlabel('Region')
plt.ylabel('Number of Outlets')
plt.legend(title='Outlet Type')
plt.show()
```

## How many unique employees are assigned to each warehouse?

```python
pvt_employees = pd.pivot_table(
    outletsFinal,
    values='Employee_Code',
    index='Warehouse Name',
    aggfunc='nunique'
)
pvt_employees.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Number of Employees per Warehouse')
plt.xlabel('Warehouse Name')
plt.ylabel('Number of Employees')
plt.show()
```

```python
pvt_outlet_types = pd.pivot_table(
    outletsFinal,
    values='OutletId',
    index='Region',
    columns='Outlet Type',
    aggfunc='count',
    fill_value=0
)

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(pvt_outlet_types, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Outlet Types by Region')
plt.xlabel('Outlet Type')
plt.ylabel('Region')
plt.show()
```
